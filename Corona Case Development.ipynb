```python
most_confirmed = confirmed.groupby('country').sum()
sort_countries = most_confirmed.sort_values(by='confirmed', ascending=False)
top_countries = sort_countries[0:9]
top_countries 
```

```python
top_countries_index = top_countries.index
```

```python
top_countries_index
```

```python
time_country_confirmed = confirmed.pivot(index='time', columns='country', values='confirmed')
time_country_confirmed.fillna(0)
```

```python
confirmed_list = []
#looping over the countries, starting with empty list

for country in top_countries_index:
    confirmed_list.append(time_country_confirmed[country].fillna(0).values)
    
dates =time_country_confirmed.index
country_names= 'Turkey, Iran, UK, France, Germany, China, Spain, Italy, US'

plt.figure(figsize=(13,10))
plt.style.use('seaborn')
plt.stackplot(dates, confirmed_list, labels=country_names)
plt.xticks(np.arange(125, step=30), ('Jan', 'Feb', 'Mar', 'Apr', 'May'))
plt.xlabel('time', size=15)
plt.ylabel('# of cases', size=15)
plt.title('WatchMojo: Top 9 COVID19 countries', color='royalblue', size=20)

plt.legend(['Turkey', 'Iran', 'UK', 'France', 'Germany', 'China', 'Spain', 'Italy', 'US'], loc='upper left', fontsize='large')

plt.show()
```

```python
deaths = pd.read_csv('../data/deaths.csv')
recovered = pd.read_csv('../data/recovered.csv')
```

```python
deaths = deaths.fillna(0)
recovered = recovered.fillna(0)
```

```python
deaths.info()
deaths.describe()
```

```python
recovered.info()
recovered.describe()
```

```python
merged_df0 = pd.merge(confirmed.fillna(0), deaths, on='id')
merged_df =pd.merge(merged_df0, recovered, on='id')
```

```python
merged_df = merged_df.fillna(0)
print(merged_df)
```

```python
merged_df['netinfected'] = merged_df['confirmed'] - merged_df['recovered'] - merged_df['deaths']
```

```python
merged_df
```

```python
#reshaping the deaths and recovered columns to negative numbers
merged_df[['deaths', 'recovered']]=merged_df[['deaths', 'recovered']]*(-1)
```

```python
merged_df
```

```python
#plotting for every single country: first look at the specific properties for the dataframe of one country 
#example GERMANY

merged_GER = merged_df[merged_df.country=='Germany']
merged_GER
```

```python
#now making TIME the index of the dataframe and accumulate the single variables over time 
merged_GER_time=merged_GER.groupby('time').sum()
merged_GER_time
```

```python
#now excluding every single column we actually don't need 

merged_GER_final = merged_GER_time[['confirmed', 'deaths', 'recovered', 'netinfected']]
merged_GER_final.tail()
```

```python
#plotting for one country as example
#we have to consider the order from highest to lowest
#functions will be plotted in layers
fig=plt.figure(figsize=(12, 12))
plt.xticks(np.arange(130, step=30), ('Jan', 'Feb', 'Mar', 'Apr'))
plt.fill_between(merged_GER_final.index,merged_GER_final['confirmed'], color='royalblue')
plt.fill_between(merged_GER_final.index,merged_GER_final['recovered'], color='lime')
plt.fill_between(merged_GER_final.index,merged_GER_final['deaths'], color='r')
plt.fill_between(merged_GER_final.index,merged_GER_final['netinfected'], color='black')
plt.xlabel('Date')
plt.ylabel('Corona in Numbers')
plt.legend(['confirmed', 'recovered', 'deaths', 'netinfected'], loc='upper left')
```

```python
#now for all countries done in a loop

fig=plt.figure(figsize=(20, 20))

for i in range(0,9):
    country=merged_df[merged_df.country==top_countries.index[i]]
    country_time=country.groupby('time').sum()
    country_final=country_time[['confirmed', 'recovered', 'deaths', 'netinfected']]
    
    plt.subplot(3,3,i+1)
    plt.xticks(np.arange(130, step=30), ('Jan', 'Feb', 'Mar', 'Apr'))
    plotdata = country_final
    plt.plot(plotdata)
    
    plt.title(top_countries.index[i], size=20)
    plt.fill_between(country_final.index,country_final['confirmed'], color='royalblue')
    plt.fill_between(country_final.index,country_final['recovered'], color='lime')
    plt.fill_between(country_final.index,country_final['deaths'], color='red')
    plt.fill_between(country_final.index,country_final['netinfected'], color='black')

    
fig.autofmt_xdate()
plt.subplots_adjust(top=0.935, left=0.075)
fig.text(0.5, 0.17,'Date', size= 18)
fig.text(0, 0.652, 'How Patient Conditions changed', rotation=90, size=18)
fig.suptitle('COVID: Patients/Country and their course of disease', color='royalblue', size=25)
plt.legend( ['confirmed', 'recovered', 'deaths', 'netinfected'], bbox_to_anchor=(0.05, -0.25), 
               fancybox=True, shadow=True, ncol=4, fontsize='large')

    
```
